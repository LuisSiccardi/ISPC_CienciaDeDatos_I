```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el dataset
encuestaSalarial = pd.read_csv('Datos/EncuestaRemuneracionSalarialArgentina2023.csv')

# Seleccionar las 10 columnas importantes
columnas_importantes = [
    'donde_estas_trabajando', 'dedicacion', 'tipo_de_contrato',
    'ultimo_salario_mensual_o_retiro_bruto_en_tu_moneda_local', 'pagos_en_dolares',
    'que_porcentaje_o_monto_recibis_en_criptomonedas', 'recibis_algun_tipo_de_bono',
    'como_consideras_que_estan_tus_ingresos_laborales_comparados_con_el_semestre_anterior',
    'trabajo_de', 'anos_de_experiencia'
]

# Filtrar el DataFrame
encuestaSalarial = encuestaSalarial[columnas_importantes]

# Dividir las columnas en numéricas y categóricas
columnas_numericas = encuestaSalarial.select_dtypes(include=[np.number]).columns
columnas_categoricas = encuestaSalarial.select_dtypes(include=['object']).columns

# Medidas descriptivas y gráficos para variables numéricas
for columna in columnas_numericas:
    media = encuestaSalarial[columna].mean()
    mediana = encuestaSalarial[columna].median()
    desviacion_estandar = encuestaSalarial[columna].std()
    skewness = encuestaSalarial[columna].skew()

    # Histograma
    plt.figure(figsize=(10, 6))
    plt.hist(encuestaSalarial[columna].dropna(), bins=20, color='lightblue', edgecolor='black')
    plt.xlabel(columna)
    plt.ylabel('Frecuencia')
    plt.title(f'Distribución de {columna}')
    plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label='Media')
    plt.axvline(mediana, color='green', linestyle='dashed', linewidth=1, label='Mediana')
    plt.legend()
    plt.show()

    print(f"Variable numérica: {columna}")
    print(f"Media: {media:.2f}")
    print(f"Mediana: {mediana}")
    print(f"Desviación Estándar: {desviacion_estandar:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print()

# Gráficos para variables categóricas
for columna in columnas_categoricas:
    conteo_categorias = encuestaSalarial[columna].value_counts()
    proporciones = encuestaSalarial[columna].value_counts(normalize=True) * 100

    # Agrupar categorías menos frecuentes en "Otros"
    n_categorias_mostrar = 15
    conteo_categorias_otros = conteo_categorias.head(n_categorias_mostrar)
    if len(conteo_categorias) > n_categorias_mostrar:
        otras_categorias = conteo_categorias.iloc[n_categorias_mostrar:].sum()
        conteo_categorias_otros['Otros'] = otras_categorias

    # Visualizar el gráfico con la nueva categoría "Otros"
    plt.figure(figsize=(10, 6))
    ax = conteo_categorias_otros.plot(kind='bar', color='skyblue')
    plt.xlabel(columna)
    plt.ylabel('Cantidad')
    plt.title(f'Distribución de {columna}')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print(f"Variable categórica: {columna}")
    print("Conteo de cada categoría:")
    print(conteo_categorias)
    print("Proporción de cada categoría (%):")
    print(proporciones)
    print()
```

```python
import pandas as pd
import numpy as np

# lectura del dataset
encuestaSalarial = pd.read_csv('Datos/EncuestaRemuneracionSalarialArgentina2023.csv')
#print("Dataset original")
#display(encuestaSalarial.columns)
```

### Extracción y limpieza de datos

```python
# Convierto los valores en nros

datos=encuestaSalarial[['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local','anos_de_experiencia']].copy()

# Convierte valores en float

datos['anos_de_experiencia'] = (pd.to_numeric(datos['anos_de_experiencia'], errors='coerce'))
datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'] = pd.to_numeric(datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'], errors='coerce')

# Limita el dataset

datos = datos[datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'] < 1000000] # Limita el dataset a un rango de ingresos menores a 1 millón
datos = datos[datos['anos_de_experiencia'] < 20] # Limita el dataset a un rengo de 20 años de experiencia

datos.loc[datos['anos_de_experiencia']==0,'anos_de_experiencia']=0.5 # Se modifica ingreso 0 a 0.5 para que tenga un valor
datos = datos.drop(datos.index[-1])

display(encuestaSalarial.head(10))
```

```python

numero_columnas = datos.select_dtypes(include=[np.number]).columns
for columna in numero_columnas:
    print(f"Analisis de la columna: {columna}")

    total_valores = datos[columna].size
    valores_validos= datos[columna].count()
    porcentaje = valores_validos / total_valores * 100


    valor_medio = datos[columna].mean()
    mediana = datos[columna].median()

    desv_std = datos[columna].std()

    valores_ausentes = datos[columna].isnull().sum()

    valores_erroneos = datos[columna][datos[columna] < 0].count()

    print(f"Total de valores: {total_valores}")
    print(f"Valores válidos: {valores_validos}")
    print(f"Porsentaje de valores validos: {porcentaje:.2f}%")
    print(f"Media: {valor_medio}")
    print(f"Mediana: {mediana}")
    print(f"Desviación estándar: {desv_std}")
    print(f"Valores ausentes: {valores_ausentes}")
    print(f"Valores erróneos: {valores_erroneos}")
    print("\n")


display(datos)
```

```python
# Elimina filas con datos nulos

datos = datos.dropna(subset=['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'])
datos = datos.reset_index(drop=True)
datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'] = datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'].astype(int)
display(datos)
```

```python
import matplotlib.pyplot as plt

plt.ylabel("Último salario mensual - $")
plt.xlabel('Años de experiencia')
plt.scatter(datos['anos_de_experiencia'], datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'], color="green")
plt.show()
```

```python
from sklearn import linear_model
regresion = linear_model.LinearRegression()
anios = datos["anos_de_experiencia"].values.reshape((-1, 1))
modelo = regresion.fit(anios, datos["ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local"])
entrada = [[13.0], [8.0], [15.0],[5.0]]
modelo.predict(entrada)
plt.scatter(entrada, modelo.predict(entrada), color="red") #datos de prueba
plt.plot(entrada, modelo.predict(entrada), color="black") #recta

plt.ylabel("Salario ($)")
plt.xlabel("Promedio de años de experiencia")
plt.scatter(datos["anos_de_experiencia"], datos["ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local"], color="green") #datos de entrenamiento
plt.show()

# Mostramos el b y a de nuestra recta
print("Intersección (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

### RSME, MAE y R2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from scipy.stats import pearsonr  # Importar Pearson

# Convertir los valores a números
datos = encuestaSalarial[['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local', 'anos_de_experiencia']].copy()
datos['anos_de_experiencia'] = pd.to_numeric(datos['anos_de_experiencia'], errors='coerce')
datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'] = pd.to_numeric(datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'], errors='coerce')

# Filtros para limitar los valores
datos = datos[datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'] < 1000000]
datos = datos[datos['anos_de_experiencia'] < 20]

# Separar variables independientes y dependientes
X = datos["anos_de_experiencia"].values.reshape(-1, 1)  # Años de experiencia
y = datos["ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local"]  # Salario

# Crear el modelo de regresión lineal
regresion = linear_model.LinearRegression()
modelo = regresion.fit(X, y)

# Datos de prueba
entrada = [[13.0], [8.0], [15.0], [5.0]]
predicciones = modelo.predict(entrada)

# Visualización
plt.scatter(entrada, predicciones, color="red")  # Datos de prueba
plt.plot(X, modelo.predict(X), color="black")  # Línea de regresión
plt.ylabel("Salario ($)")
plt.xlabel("Promedio de años de experiencia")
plt.scatter(datos["anos_de_experiencia"], datos["ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local"], color="green")  # Datos de entrenamiento
plt.show()

# Mostramos el intercepto y la pendiente de la recta
print("Intersección (b):", modelo.intercept_)
print("Pendiente (m):", modelo.coef_)

# Calcular métricas de rendimiento
y_pred = modelo.predict(X)  # Predicciones para los datos de entrenamiento

# RMSE
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f'RMSE: {rmse}')

# MAE
mae = mean_absolute_error(y, y_pred)
print(f'MAE: {mae}')

# R²
r2 = r2_score(y, y_pred)
print(f'R²: {r2}')

# Coeficiente de Pearson
pearson_corr, _ = pearsonr(datos['anos_de_experiencia'], datos['ultimo_salario_mensual_o_retiro_neto_en_tu_moneda_local'])
print(f'Coeficiente de correlación de Pearson: {pearson_corr}')
```

### Concluciones :

1.   Alto Error Cuadrático Medio (RMSE):
